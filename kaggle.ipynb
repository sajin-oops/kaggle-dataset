```python
# 1. Ensure 'streams' column is of string type before applying string operations
df['streams'] = df['streams'].astype(str)

# 2. Clean 'streams' column by removing any non-numeric values
df = df[df['streams'].str.isnumeric()]

# 3. Now, convert 'streams' to integers using 'Int64' which can handle large numbers and missing values
df['streams'] = df['streams'].replace(',', '', regex=True).astype('Int64')

# 4. Clean 'in_deezer_playlists' by removing commas and converting to integers
df['in_deezer_playlists'] = df['in_deezer_playlists'].replace(',', '', regex=True).replace('Not Found', 0).astype('Int64')

# 5. Handle 'in_shazam_charts' by converting to integers
df['in_shazam_charts'] = pd.to_numeric(df['in_shazam_charts'], errors='coerce').fillna(0).astype(int)

# 6. For columns like 'key' with missing values, fill with 'Unknown'
df['key'] = df['key'].fillna('Unknown')

# 7. Check and handle any duplicate rows
df = df.drop_duplicates()

# 8. Strip any extra whitespace from string columns
str_columns = df.select_dtypes(include=['object']).columns
df[str_columns] = df[str_columns].apply(lambda x: x.str.strip())

# 9. Basic outlier handling (if desired)
df = df[df['streams'] < 10_000_000_000]

# Display cleaned data summary
df.info(), df.head()
```

```python
top_artists = df.groupby('artist(s)_name')['streams'].sum().nlargest(10).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(x='streams', y='artist(s)_name', data=top_artists, palette='Blues_d')
plt.title('Top 10 Artists by Stream Count')
plt.xlabel('Total Streams')
plt.ylabel('Artists')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['danceability_%'], bins=20, kde=True, color='purple')
plt.title('Distribution of Danceability')
plt.xlabel('Danceability (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='danceability_%', y='energy_%', data=df, hue='artist(s)_name', legend=False)
plt.title('Danceability vs Energy')
plt.xlabel('Danceability (%)')
plt.ylabel('Energy (%)')
plt.show()
```

```python
top_tracks = df.nlargest(10, 'in_spotify_playlists')[['track_name', 'in_spotify_playlists']]

plt.figure(figsize=(10,6))
sns.barplot(x='in_spotify_playlists', y='track_name', data=top_tracks, palette='Greens_d')
plt.title('Top 10 Tracks by Spotify Playlists Count')
plt.xlabel('Number of Spotify Playlists')
plt.ylabel('Track Name')
plt.show()
```

```python
streams_by_year = df.groupby('released_year')['streams'].sum().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(x='released_year', y='streams', data=streams_by_year, marker='o')
plt.title('Total Streams by Released Year')
plt.xlabel('Year')
plt.ylabel('Total Streams')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[['streams', 'danceability_%', 'energy_%', 'valence_%', 'acousticness_%', 'speechiness_%']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Audio Features')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='valence_%', y='energy_%', size='danceability_%', data=df, hue='artist(s)_name', legend=False, sizes=(20, 200))
plt.title('Valence vs Energy with Danceability')
plt.xlabel('Valence (%)')
plt.ylabel('Energy (%)')
plt.show()
```
